```python
clicks_train['night'] = clicks_train
```

```python
clicks_train['weekend'] = clicks_train['week_day'].apply(lambda x:1 if x in ['Friday','Saturday','Sunday'] else 0)
```

```python
clicks_train['weekday'] = clicks_train['weekend'].apply(lambda x:1 if x == 0 else 0)
```

```python
clicks_train_temp = clicks_train[['account_id','day','night','weekend','weekday']]
```

```python
clicks_train_temp
```

```python
add_activity.reset_index(inplace=True)
```

```python
add_activity
```

```python
new_users_2
```

```python
new_users_2['day_ratio'] = new_users_2['day'] / new_users_2['total_clicks']
```

```python
new_users_2['night_ratio'] = new_users_2['night'] / new_users_2['total_clicks']
new_users_2['weekend_ratio'] = new_users_2['weekend'] / new_users_2['total_clicks']
new_users_2['weekday_ratio'] = new_users_2['weekday'] / new_users_2['total_clicks']
new_users_2['involvement'] = new_users_2['total_clicks'] / new_users_2['order_counts']
```
